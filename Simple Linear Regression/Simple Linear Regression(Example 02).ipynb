```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
from google.colab import files
uploaded = files.upload()

dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
years_exp = dataset['YearsExperience']
salary = dataset['Salary']

print(years_exp)
print(salary)
```

```python
plt.scatter(years_exp, salary)
plt.show()
```

```python
X = np.array(years_exp).reshape(-1,1)
y = np.array(salary).reshape(-1,1)
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
regression_model_mse = mean_squared_error(X,y)
print('Mean Squared Error:\t',math.sqrt(regression_model_mse))
print("R squared value\t\t",model.score(X,y))
```

```python
print(model.coef_[0])
print(model.intercept_[0])
```

```python
plt.scatter(X,y, color = 'orange')
plt.plot(X,model.predict(X), color = 'black')
plt.title('Linear Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
print('Prediction by the model:\t', model.predict([[8.2]]))
```
